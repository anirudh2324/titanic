```python
import pickle
```

```python

test_model=pickle.load(open('final_model_titanic_jul.pk','rb'))
```

```python
import pandas as pd
import numpy as np
```

```python
test_data=pd.read_csv('test.csv')
```

```python
test_data.head()
```

```python
test_data.Age =test_data.Age.fillna(np.mean(test_data.Age))
test_data.Fare =test_data.Fare.fillna(np.mean(test_data.Fare))
```

```python
test_data.isnull().sum()
```

```python
test_data= pd.get_dummies(test_data,columns=['Pclass','Sex','Embarked'])
test_data.head()
```

```python
test_data.columns
```

```python
test_data.drop('Ticket',axis=1,inplace=True)

test_data.head()
```

```python
test_data.drop('Name',axis=1,inplace=True)

test_data.head()
```

```python
test_data.drop('Cabin',axis=1,inplace=True)
test_data.head()
```

```python
new_pred=test_model.predict(test_data)
```

```python
test_data['servived']=new_pred
```

```python
test_data.head()
```

```python
test_data.to_excel('Final_test_titanic_prediction.xlsx')
```
